```python
import pickle
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
timess = [None]*3

for i in range(2+1):
    s = f"large_times_{i}.txt"
    if not os.path.isfile(s):
        print("File " + s + " does not exist!")
        continue

    with open(s, "rb") as f:
        times = pickle.load(f)
    print(times[:3])
    timess[i] = np.array(times)
```

```python
primes = []

with open("primes50.txt") as f:
    primes = [int(x) for x in f.read().strip("\n").split()]

print(primes[:3])
```

```python
print(timess[2].shape)
slice = 1000
plt.errorbar(primes[:slice], timess[2][:slice, 0], yerr=timess[2][:slice, 1], ls='none', capsize=3, fmt=".", markersize=10)
plt.errorbar(primes[:slice], timess[1][:slice, 0], yerr=timess[1][:slice, 1], ls='none', capsize=3, fmt=".", markersize=10)
plt.errorbar(primes[:slice], timess[0][:slice, 0], yerr=timess[0][:slice, 1], ls='none', capsize=3, fmt=".", markersize=10)
plt.show()
```
