```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("bench8.txt").read()
runs = f.split("Algorithm")[1:]
# print(len(runs), runs[0])


trial = [float(x.split()[2][:-1]) for x in  runs[0].split("\n")[3:][:-1]]
mr = [float(x.split()[2][:-1]) for x in  runs[1].split("\n")[3:][:-1]]
pse = [float(x.split()[2][:-1]) for x in  runs[2].split("\n")[3:][:-1]]

print(len(trial))
print(len(mr))
```

```python
# This follows from main.cpp
maxInt = 20000000
wsize = 10000
xs = [1000 * i for i in range(maxInt//wsize)]
```

```python
plt.scatter(xs, trial, s=1)
plt.show()
```

```python
plt.scatter(xs, mr, s=1)
plt.show()
```

```python
plt.scatter(xs, pse, s=1)
plt.show()
```

```python
plt.scatter(xs[10:], 1000*np.array(trial)[10:], s=1, label="Trial division")
plt.scatter(xs[10:], 1000*np.array(mr)[10:], s=1, label="Miller-Rabin")
plt.scatter(xs[10:], 1000*np.array(pse)[10:], s=1, label="BailliePSW")

plt.ylabel("Time [ms]")

plt.legend()
plt.show()
```
