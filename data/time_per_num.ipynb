```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = open("bench1.txt").read()
runs = f.split("Algorithm")[1:]
# print(len(runs), runs[0])


trial = [float(x.split()[2][:-1]) for x in  runs[0].split("\n")[3:][:-1]]
mr = [float(x.split()[2][:-1]) for x in  runs[1].split("\n")[3:][:-1]]
pse = [float(x.split()[2][:-1]) for x in  runs[2].split("\n")[3:][:-1]]

print(len(trial))
print(len(mr))
```

```python
# This follows from main.cpp
maxInt = 2037035976334486086268445688409378161051468393665936250636140449354381299763336706183397376  # 2**300
logMaxInt = 300

xs = []
i = 1
while i <= logMaxInt:
    xs.append(i)
    i += 1
print(len(xs))
```

```python
plt.scatter(xs, trial, s=1)
plt.show()
```

```python
plt.scatter(xs, mr, s=1)
plt.show()
```

```python
plt.scatter(xs, pse, s=1)
plt.show()
```

```python
# to us div repeats == 1e6 / 10 == 1e5
plt.scatter(xs[10:], 10**5*np.array(trial)[10:], s=1, label="Trial division")
plt.scatter(xs[10:], 10**5*np.array(mr)[10:], s=1, label="Miller-Rabin")
plt.scatter(xs[10:], 10**5*np.array(pse)[10:], s=1, label="BailliePSW")

plt.ylabel(r"Time [$\mu$s]")
plt.xlabel(r"$\log_2 \mathbb{N}_0$")

plt.legend()
plt.savefig("times.pdf")
plt.show()
```
